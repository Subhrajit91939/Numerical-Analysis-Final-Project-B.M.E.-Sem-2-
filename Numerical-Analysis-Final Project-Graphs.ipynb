```python
#using the .csv files to generate Graphs using Matplotlib
#importing all the Important libraries for the Graphs/Figures
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#making the Backgrounds of Plots White with Grids for Easier Interpretation.
plt.style.use('seaborn-whitegrid')


#ignore all the warnings for now
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing all the .csv files as storing them as Dataframes for Easier Access
euler = pd.read_csv('euler_output.csv', header=None, names=['X', 'Y'])
rungekutta2 = pd.read_csv('rk2_output.csv', header=None, names=['X', 'Y'])
rungekutta4 = pd.read_csv('rk4_output.csv', header=None, names=['X', 'Y'])
```

```python
# Verifiying the Total no. of : (rows,columns)
print('Euler :', euler.shape, end='\n')
print('Runge-Kutta-2 :', rungekutta2.shape, end='\n')
print('Runge-Kutta-4 :', rungekutta4.shape, end='\n')
```

```python
#Verifying Last 20 Columns for any empty Data
euler.tail(20)
```

```python
#Delete those last 2 unnecessary rows to remove any inconsistency in the graph
euler.drop(euler.tail(2).index,inplace=True)
rungekutta2.drop(rungekutta2.tail(2).index,inplace=True)
rungekutta4.drop(rungekutta4.tail(2).index,inplace=True)
```

```python
#Graph-1: Comparison between Trajectory using Euler and Runge-Kutta Methods. for $\Delta$x = 0.001
#theta = 30 deg ; v0 = 2 m/s
hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(16,5))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(euler.X, euler.Y, label="Euler-Method", color='#ff6361', linewidth=2.5)
plt.plot(rungekutta2.X, rungekutta2.Y, label="Runge-Kutta-2-Method", color='#bc5090', linewidth=2.5)
plt.plot(rungekutta4.X, rungekutta4.Y, label="Runge-Kutta-4-Method", color='#003f5c', linewidth=2.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparing Trajectory using Euler and Runge-Kutta Methods.", fontsize = 18, **hfont)

plt.text(0.3, 0.001,'$\Delta$x = 0.001\n$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=15)
plt.legend(fontsize = 14)

#Observations: All three methods are pretty much Accurate with respect to each other.
#(Euler-method differs from RK by a marginal difference).
```

```python
#Graph-2 : to Observe the difference between the different methods more accurately
#(Refer: Graph for more info...)

hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(12,10))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(euler.X, euler.Y, label="Euler-Method", color='#ff6361', linewidth=2.5)
plt.plot(rungekutta2.X, rungekutta2.Y, label="Runge-Kutta-2-Method", color='#bc5090', linewidth=2.5)
plt.plot(rungekutta4.X, rungekutta4.Y, label="Runge-Kutta-4-Method", color='#003f5c', linewidth=2.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparing Trajectory using Euler and Runge-Kutta Methods.", fontsize = 18, **hfont)

plt.text(0.3, 0.001,'$\Delta$x = 0.001\n$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=15)
plt.legend(fontsize = 14)
```

```python
eulera = pd.read_csv('euler_output1.csv', header=None, names=['X', 'Y'])
rungekutta2a = pd.read_csv('rk2_output1.csv', header=None, names=['X', 'Y'])
rungekutta4a = pd.read_csv('rk4_output1.csv', header=None, names=['X', 'Y'])
```

```python
print('Euler :', eulera.shape, end='\n')
print('Runge-Kutta-2 :', rungekutta2a.shape, end='\n')
print('Runge-Kutta-4 :', rungekutta4a.shape, end='\n')
```

```python
#Graph-3: Comparison between Trajectory using Euler and Runge-Kutta Methods. for $\Delta$x = 0.01
#theta = 30 deg ; v0 = 2 m/s

hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(16,5))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(eulera.X, eulera.Y, label="Euler-Method", color='#ff6361', linewidth=2.5)
plt.plot(rungekutta2a.X, rungekutta2a.Y, label="Runge-Kutta-2-Method", color='#bc5090', linewidth=2.5)
plt.plot(rungekutta4a.X, rungekutta4a.Y, label="Runge-Kutta-4-Method", color='#003f5c', linewidth=2.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparing Trajectory using Euler and Runge-Kutta Methods.", fontsize = 18, **hfont)

plt.text(0.365, 0.0255,'$\Delta$x = 0.01\n$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=14)
plt.legend(fontsize = 14)

#Observations: Range for Runge Kutta Methods are Nearly the Same, whereas Euler Method Overshoots 
#the actual value of Range.
```

```python
#Graph-4: Comparison between Euler-method for values : h=0.01 & h=0.001
hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(16,5))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(eulera.X, eulera.Y, label="$\Delta$x = 0.01\n", color='#ff6361', linewidth=3.5)
plt.plot(euler.X, euler.Y, label="$\Delta$x = 0.001\n", color='#003f5c', linewidth=3.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparison between Euler methods for values : h=0.01 & h=0.001", fontsize = 18, **hfont)

plt.text(0.370, 0.03,'$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=15)
plt.legend(fontsize = 14)

#Observations: Euler method(h=0.001) is more accurate than the Euler-method(h=0.01)
#Clearly indicating that Smaller the Step Size(h) the more accurate the results will be.
```

```python
#Graph-5: Comparison between Runge-Kutta-2 method for values : h=0.01 & h=0.001
hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(16,5))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(rungekutta2a.X, rungekutta2a.Y, label="$\Delta$x = 0.01\n", color='#ff6361', linewidth=3.5)
plt.plot(rungekutta2.X, rungekutta2.Y, label="$\Delta$x = 0.001\n", color='#003f5c', linewidth=3.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparison between  Runge-Kutta-2 method for values : h=0.01 & h=0.001", fontsize = 18, **hfont)

plt.text(0.370, 0.03,'$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=15)
plt.legend(fontsize = 14)

#Observation: RK2 is very Accurate for both the values of h=0.01 & h=0.001
```

```python
#Graph-6: Comparison between Runge-Kutta-4 method for values : h=0.01 & h=0.001
hfont = {'fontname':'Montserrat', 'weight' : 'medium'}
plt.figure(figsize=(16,5))

plt.rc('font', family='Montserrat', weight='medium')

plt.plot(rungekutta4a.X, rungekutta4a.Y, label="$\Delta$x = 0.01\n", color='#ff6361', linewidth=3.5)
plt.plot(rungekutta4.X, rungekutta4.Y, label="$\Delta$x = 0.001\n", color='#003f5c', linewidth=3.5)

plt.xlabel('x-co-ordinate(m)', fontsize = 18, **hfont) 
plt.ylabel('y-co-ordinate(m)',fontsize = 18, **hfont)
plt.title("Comparison between Trajectories using Runge-Kutta-4 method for values : h=0.01 & h=0.001", fontsize = 18, **hfont)

plt.text(0.370, 0.03,'$\\theta$ = 30\n$V_{0}$ = 2 m/s',
        verticalalignment='bottom', horizontalalignment='right',
        color='#222831', fontsize=15)
plt.legend(fontsize = 14)

#Observation: As Expected, RK4 is also very Accurate for both the values of h=0.01 & h=0.001
```
